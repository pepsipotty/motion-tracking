```python
import skimage 
from skimage import io 
import numpy as np 
import matplotlib.pyplot as plt
import math
import cv2
import scipy.ndimage
from scipy.linalg import eigh
import matplotlib.patches as patches
```

### Covarience Tracking

- Load Image

```python
im = skimage.io.imread('Dataset/image0.jpg')
plt.imshow(im)
plt.show()
print(im.shape)
```

```python
target = im[350:415, 290:360]

# Save the selected region as a new image
skimage.io.imsave('target.jpg', target)

# Display the selected region
plt.imshow(target)
plt.show()

# Print the shape of the selected region
print(target.shape)
```

```python
features = []

# Iterate over each pixel in the selected region
for y in range(target.shape[0]):
    for x in range(target.shape[1]):
        # Get the RGB values of the pixel
        r, g, b = target[y, x]

        # Append the feature vector [x, y, R, G, B] for this pixel
        features.append([x, y, r, g, b])  # Adding 300 and 100 to x and y to get original image coordinates

# Convert the list of features into a numpy array
features_array = np.array(features)

# Compute the covariance matrix
cov_matrix = np.cov(features_array.T)  # Transpose to get features as columns

print("Covariance Matrix:")
print(cov_matrix)
model_cov = cov_matrix
```

```python
height, width, _ = im.shape
window_rows, window_cols = 65, 70
y, x = np.arange(window_rows).astype(np.float32), np.arange(window_cols).astype(np.float32)
yv, xv = np.meshgrid(y, x, indexing='ij')
yv, xv  = yv.flatten()[np.newaxis, :], xv.flatten()[np.newaxis, :]

R, G, B = im[:, :, 0].astype(np.float32), im[:, :, 1].astype(np.float32), im[:, :, 2].astype(np.float32)

dists = np.zeros((height - window_rows, width - window_cols), dtype=np.float32)

for i in range(height - window_rows):
    for j in range(width - window_cols):
        feat = np.vstack((xv, yv,
                          R[i:i+window_rows, j:j+window_cols].flatten()[np.newaxis, :],
                          G[i:i+window_rows, j:j+window_cols].flatten()[np.newaxis, :],
                          B[i:i+window_rows, j:j+window_cols].flatten()[np.newaxis, :]))
        candidate_cov = np.cov(feat, bias=True)
        eigvals = eigh(model_cov, candidate_cov, eigvals_only=True)
        dist = np.sqrt((np.log(eigvals)**2).sum())
        dists[i, j] = dist
dists.shape
```

```python
plt.imshow(dists, cmap="gray")
plt.show()
```

```python
minDiff = dists.argmin()
minpos = minDiff // (width - window_cols), minDiff % (width - window_cols)
minpos[0], minpos[1]
```

```python
target = im[minpos[0]:minpos[0] + window_rows, minpos[1]:minpos[1] + window_cols]
plt.imshow(target)
plt.show()
```

```python
fig,ax = plt.subplots()
ax.imshow(im)
rect = patches.Rectangle((minpos[1],minpos[0]),window_cols,window_rows,linewidth=1,edgecolor='r',facecolor='none')
ax.add_patch(rect)
plt.show()
```

```python
def covarience_tracking(im, target):
    features = []

    # Build covariance matrix
    for y in range(target.shape[0]):
        for x in range(target.shape[1]):
            r, g, b = target[y, x]
            features.append([x, y, r, g, b]) 

    features_array = np.array(features)
    cov_matrix = np.cov(features_array.T) 

#     print("Covariance Matrix:")
#     print(cov_matrix)
    model_cov = cov_matrix
    
    
    # Covariance tracking
    height, width, _ = im.shape
    window_rows, window_cols,_ = target.shape
    y, x = np.arange(window_rows).astype(np.float32), np.arange(window_cols).astype(np.float32)
    yv, xv = np.meshgrid(y, x, indexing='ij')
    yv, xv  = yv.flatten()[np.newaxis, :], xv.flatten()[np.newaxis, :]

    R, G, B = im[:, :, 0].astype(np.float32), im[:, :, 1].astype(np.float32), im[:, :, 2].astype(np.float32)

    # Calculate and save the match distance
    dists = np.zeros((height - window_rows, width - window_cols), dtype=np.float32)
    for i in range(height - window_rows):
        for j in range(width - window_cols):
            feat = np.vstack((xv, yv,
                              R[i:i+window_rows, j:j+window_cols].flatten()[np.newaxis, :],
                              G[i:i+window_rows, j:j+window_cols].flatten()[np.newaxis, :],
                              B[i:i+window_rows, j:j+window_cols].flatten()[np.newaxis, :]))
            candidate_cov = np.cov(feat, bias=True)
            eigvals = eigh(model_cov, candidate_cov, eigvals_only=True)
            dist = np.sqrt((np.log(eigvals)**2).sum())
            dists[i, j] = dist
#     print(dists.shape)
    
    minDiff = dists.argmin()
    minpos = minDiff // (width - window_cols), minDiff % (width - window_cols)
#     print(minpos[0], minpos[1])
    
    # Plot the best match box
    fig,ax = plt.subplots()
    ax.imshow(im)
    rect = patches.Rectangle((minpos[1],minpos[0]),window_cols,window_rows,linewidth=1,edgecolor='r',facecolor='none')
    ax.add_patch(rect)
    plt.show()
```

```python
covarience_tracking(im, target)
```
